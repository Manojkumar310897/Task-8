```python
# It takes so much time for translating the entire dataset,so we are taking only first 100 words or sentences.
df1=df.head(100)
df1
```

```python
from deep_translator import GoogleTranslator
german =[]
for i in df1['English words/sentences']:
      a = GoogleTranslator(source='auto', target='de').translate(i)
      german.append(''.join(a))
     
```

```python
ge =pd.Series(german)
ger=pd.DataFrame(ge, columns=["german words"])
ger.head(10)
```

```python
spanish =[]
for j in df1['English words/sentences']:
  b = GoogleTranslator(source='auto', target='spanish').translate(j)
  spanish.append(''.join(b)) 
```

```python
sp =pd.Series(spanish)
spa=pd.DataFrame(sp, columns=["spanish words"])
spa.head(10)    
```

```python
french =[]
for k in df1['English words/sentences']:
  c = GoogleTranslator(source='auto', target='french').translate(k)
  french.append(''.join(c))
```

```python
fe =pd.Series(french)
fer=pd.DataFrame(fe, columns=["french words"])
fer.head(10)    
```

```python
different_lan = pd.concat([df1,ger,spa,fer],axis=1)
different_lan.head(10)
```
