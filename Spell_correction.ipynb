```python
from textblob import Word
correct_word=[]
for i in df2["English words/sentences"]:
  word=Word(str(i))
  word_correct=word.correct()
  correct_word.append(word_correct)
print(correct_word)
```

```python
crt=pd.Series(correct_word)
print(crt)
```

```python
words=pd.DataFrame(crt,columns=["Corrected English Words"])
words.head()
```

```python
from textblob import Word   
from textblob import TextBlob
def correct_sentence_spelling(sentence):
    sentence = TextBlob(sentence)
    result = sentence.correct()
    return result


txt_data =df3['English words/sentences']

crt_txt_data=[]
for i in txt_data:
    if len(i) > 15:
        a = correct_sentence_spelling(i)
        crt_txt_data.append(a)
       
crt_txt_data
```
